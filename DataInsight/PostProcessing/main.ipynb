```python
labels_list = ["NAME","AGE","GENDER","DATE","DIAGNOSIS","HISTORY",
               "BP","TEMP","WEIGHT","MEDICINE TYPE",
               "MEDICINE NAME","MEDICINE POWER","MEDICINE DOSE",
               "B","I","B-U","I-U","PULSE"]
```

```python
import pickle
import json

# Load the content of the pickle file
with open('D:\Softwares\Visual Studio Code\Python\DataInsight\PostProcessing\Validation_input.pkl', 'rb') as file:
    data = pickle.load(file)

# Create the new JSON structure
new_json = {
    "patient info": {
        "NAME": "",
        "AGE": "",
        "GENDER": "",
        "DATE": ""
    },
    "MEDICINE BLOCK": [],
    "DIAG BLOCK": {
        "DIAGNOSIS": " ",
        "HISTORY": " ",
        "BP": " ",
        "TEMP": " ",
        "WEIGHT": " "
    }
}

# Assign values from pickle data to the new JSON structure
for idx, label in enumerate(labels_list):
  
    if label in ["NAME", "AGE", "GENDER", "DATE"]:
  
        new_json["patient info"][label] = data[0]['text'][data[0]['label'].index(idx)]
  
    elif label == "DIAGNOSIS":
  
        new_json["DIAG BLOCK"]["DIAGNOSIS"] = data[0]['text'][data[0]['label'].index(idx)]
  
    elif label in ["BP", "TEMP", "WEIGHT"]:
  
        new_json["DIAG BLOCK"][label] = data[0]['text'][data[0]['label'].index(idx)]
  
    elif label == "MEDICINE TYPE":
  
        medicine_indices = [i for i, l in enumerate(data[0]['label']) if l == idx]
        for medicine_idx in medicine_indices:
  
            medicine = {"MEDICINE TYPE": data[0]['text'][medicine_idx]}
  
            for next_label in ["MEDICINE NAME", "MEDICINE POWER", "MEDICINE DOSE"]:
  
                if next_label in labels_list:
  
                    next_idx = labels_list.index(next_label)
  
                    if next_idx < len(data[0]['label']):
  
                        next_label_index = data[0]['label'].index(next_idx, medicine_idx)
                        medicine[next_label] = data[0]['text'][next_label_index]
  
            new_json["MEDICINE BLOCK"].append(medicine)

# Write the new JSON structure to a file
with open('new_data.json', 'w') as file:
    json.dump(new_json, file, indent=4)

print("New JSON file created successfully.")
```
